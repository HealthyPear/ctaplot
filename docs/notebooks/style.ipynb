# ctaplot set style
On can set ctaplot plot style easily with `ctaplot.set_style()`. Here is an example.

```python
import ctaplot
```

```python
# Before
ctaplot.plot_angular_resolution_cta_requirement('north')
```

```python
# After
ctaplot.set_style('slides')
ctaplot.plot_angular_resolution_cta_requirement('north')
```
